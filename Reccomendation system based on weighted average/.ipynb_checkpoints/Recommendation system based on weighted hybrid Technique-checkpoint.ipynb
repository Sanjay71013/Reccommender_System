# Basic Reccomendation System

```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head()
```

```python
print("Credits:",credits.shape)
print("Movies:",movies.shape)
```

```python
## Combining the two dataframes by id cloumn
credits_column_renamed = credits.rename(index=str,columns={'movie_id':'id'})
df = movies.merge(credits_column_renamed,on='id')
df.head()
```

```python
df = df.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
df.head()
```

```python
df.info()
```

No Null values in the df

### Using weighted average for movie's average rating (voting_average)

<img src='https://miro.medium.com/max/736/1*fGziZl2Do-VyQXSCPq_Y2Q.png'>

```python
## Calculate all the components based on the above formula
v = df['vote_count']
R = df['vote_average']
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.70)
```

```python
df['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
df.head()
```

```python
df = df.sort_values('weighted_average', ascending=False)
df[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

## Reccomendation based on weighted average

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
axis = sns.barplot(x=df['weighted_average'].head(10),y=df['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies by average votes',weight='bold')
plt.xlabel('Weighted Average Score',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.savefig('best_movies.png')
```

## Reccomandation based on popularity

```python
df_popularity = df.sort_values('popularity', ascending=False)
plt.figure(figsize=(12,6))
axis = sns.barplot(x=df_popularity['popularity'].head(10),y=df_popularity['original_title'].head(10))
plt.title('Most Popular by votes',weight='bold')
plt.xlabel('Score of Popularity',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.savefig('most_popular_movies.png')
```

## Reccomdation Based on scaled weighted average and popularity score (Priority is given 50% both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
df_scaled = scaling.fit_transform(df[['weighted_average','popularity']])
df_normalized = pd.DataFrame(df_scaled, columns=['weighted_average','popularity'])
df_normalized.head()
```

```python
df[['normalized_weight_average','normalized_popularity']] = df_normalized
df.head()
```

```python
df['score'] = df['normalized_weight_average'] *0.5 + df['normalized_popularity']*0.5
df = df.sort_values('score', ascending=False)
df[['original_title','normalized_weight_average','normalized_popularity','score']].head(10)
```

```python
plt.figure(figsize=(12,6))
axis = sns.barplot(x=df['score'].head(10),y=df['original_title'].head(10))
plt.title('Best Rated and Most popular movies',weight='bold')
plt.xlabel('Score',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.savefig('best_rated_most_popular_movie.png')
```
